# Formando um Final Data com os dados que limpamos

```python
import pandas as pd

d1 = pd.read_csv('dataset/cleaned_data/clean_data_2019.csv', sep=',')
d2 = pd.read_csv('dataset/cleaned_data/clean_data_2020.csv', sep=',')
d3 = pd.read_csv('dataset/cleaned_data/clean_data_2021.csv', sep=',')
d4 = pd.read_csv('dataset/cleaned_data/clean_data_2022.csv', sep=',')
print(f'Tamanho D1:{d1.__len__()}\nTamanho D2:{d2.__len__()}\nTamanho D3:{d3.__len__()}\nTamanho D4:{d4.__len__()}')

dataFinal = pd.concat([d1,d2,d3,d4])
print(dataFinal.__len__())

print(f'\nDados Duplicados:\nAnalise Geral-> {dataFinal.duplicated().sum()}\
    \nAnalisando de acordo com os DOC-> {dataFinal["COD_DOCUMENTO"].duplicated().sum()}')

dataFinal.to_csv('dataset/despesa_ceaps_2019-2022.csv', index = False)
print('\n\nDocumento Final Criado :)')
```
